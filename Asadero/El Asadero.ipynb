```python
# -*- coding: utf-8 -*-
import simpy
from scipy import stats

class Pozoleria:
        def __init__(self,env, mesa, mesero):
            self.mesa = simpy.Resource(env,capacity=mesa)
            self.mesero = simpy.Resource(env,capacity=mesero)

def cliente(nombre,env, pozoleria ):
    
    lst_llegadas.append(env.now)
    with pozoleria.mesa.request() as req1:
        yield req1
        yield env.timeout(abs(np.random.normal(loc=2,scale=.5)))
        
        with pozoleria.mesero.request() as req4:
            yield req4
            lst1_atencion.append(env.now)
            yield env.timeout(np.random.uniform(high=2,low=0))
            
            
        yield env.timeout(abs(stats.mielke.rvs(2.610503066037011,4.2183378080882532, 
                                              -0.32728199512873146, 3.912756509109701)))
        
        
        with pozoleria.mesero.request() as req2:
            
            yield req2
            lst_atenciones.append( env.now)
            yield env.timeout(abs(np.random.normal(loc=2.6105030660,scale=.5)))
            
        
        yield env.timeout(abs(stats.triang.rvs(0.36993648418086633, 
                                           1.022546387613386, 
                                           12.824148122239986)))
            
        with pozoleria.mesero.request() as req3:
            
            yield req3
            lst3_atencion.append(env.now)
            yield env.timeout(abs(np.random.normal(loc=2,scale=.5)))
            
         
        
        yield env.timeout(abs(stats.ncf.rvs(45.000944774399002, 3032.2299467312787, 
                                            5.3189147158760264, -13.100366512708419, 
                                            39.736777212193161))+10)


        lst_salidas.append((nombre, env.now))

def generador_clientes(env, pozoleria, cantidad):
    for i in range(cantidad):
        env.process(cliente(i,env,pozoleria))
        yield env.timeout(abs(stats.johnsonsb.rvs(1.4500376729428863, 
                                              0.099999999429103509, 
                                              1.4043133162940968)))
```

```python
%%time
z=[]
for mesas in range(24,26, 2):
    contador=1
    for mesero in range(1,11) :

        clt_nog=[]
  
        for j in range(34,52,2):
            clt_no=[]
            atn_no=[]
            atn2_no=[]
            for k in range(100):
                lst_llegadas = []
                lst_atenciones = []
                lst_salidas = []
                lst1_atencion=[]
                lst3_atencion=[]
                
                env = simpy.Environment()
                pozoleria = Pozoleria(env, mesa=mesas, mesero=mesero)
                cantidad=int(j)
                gen_clientes = env.process(generador_clientes(env,pozoleria, cantidad))
                env.run(until=120)
                clt_no.append(len(lst_llegadas)-len(lst_atenciones))

                atn_no.append(np.mean(map(lambda x,y: abs(x-y)-np.random.choice([9,8,7,6],p=[.25,.25,.25,.25]),lst_llegadas[:len(lst1_atencion)]
                                   ,lst1_atencion)))
                atn2_no.append(np.mean(map(lambda x,y: abs(x-y)-np.random.choice([5,4,7,6],p=[.25,.25,.25,.25]),
                                           lst_atenciones[:len(lst3_atencion)]
                                   ,lst3_atencion)))
                
            clt_nog.append(("Con %d clientes" %j,np.mean(clt_no)) )
            clt_nog.append(("Tiempo promedio de espera", np.mean(atn_no)) )
            clt_nog.append(("Tiempo promedio de espera 2", np.mean(atn2_no)) )
            
        d=dict(clt_nog)
        d["Meseros"]=mesero
        d["Mesas"]=mesas
        
        if contador==1: xc=pd.DataFrame(d, index=[contador])
        elif contador>1: xc=pd.concat([xc,pd.DataFrame(d, index=[mesero])])
        contador+=1
    z.append(xc)
    
```
